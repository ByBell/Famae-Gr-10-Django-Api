```python
# Extract Yes/No values and put them into another column

data['health_limit'] = data['health_limit_exceeded'].str.replace(r'(Yes|No)', '')
data['health_limit_exceeded'] = data['health_limit_exceeded'].str.extract(r'(Yes|No)*')[0]

# Same for legal_limit
data['legal_limit'] = data['legal_limit_exceeded'].str.replace(r'(Yes|No)', '')
data['legal_limit_exceeded'] = data['legal_limit_exceeded'].str.extract(r'(Yes|No)*')[0]

data
```

```python
pd.options.mode.chained_assignment = None

# Extract unit
columns = ['average_result', 'max_result', 'health_limit', 'legal_limit']

# First extract unit
for column in columns:
    data[column][data[column].str.contains('non-enforceable health')] = "0.0 ppb"
    data[column].dropna()
    
    print(data.info())
  
    splitted = data[column].str.split()
    data[column], data['unit'] = data[column].str.split(' ', 1).str
    
#data[['contaminant', 'unit', 'average_result']]
```
